```python
from typing import Dict, List, Tuple, Set, Optional, Any
import pandas as pd
import os
```

```python
model_ckpt_dict = {'bloom-560m': ['1000', '100000', '200000', '300000', '400000', '600000'], 
                   'bloom-1b1': ['1000', '10000', '100000', '200000', '300000', '400000', '500000', '600000'], 
                   'bloom-1b7': ['50000', '100000', '150000', '200000'],
                   }

xnli_result_path = os.path.join('xtreme/xnli/')
output_csv_file_path = 'csv_files/'
```

```python
def read_acc_by_ckpt(file):
    acc_dict = {}
    # acc_dict: Dict{lang:acc}
    with open(file, 'r') as f:
        lines = f.readlines()
        for i in range(1, len(lines)):
            acc_dict[lines[i].split('=')[0]] = float(lines[i].split('=')[1])
    del acc_dict['total']
    return acc_dict
```

```python
for model, ckpt_list in model_ckpt_dict.items():
    
    output_csv_file_name = os.path.join(output_csv_file_path, f'{model}_xnli_acc.csv')
    ckpt_acc_dict = {}
    # Dict{ckpt: Dict{ lang: acc}}
    for ckpt in ckpt_list:
        ckpt_acc_dict[ckpt] = read_acc_by_ckpt(os.path.join(xnli_result_path, model, f'test_results_{ckpt}.txt'))
    
    df = pd.DataFrame.from_dict(ckpt_acc_dict)
    print(df)
    df.to_csv(output_csv_file_name)
```
